<h2 style='color:#3186cc'><span style='color:red'> 1.</span> Import essential packages</h2>

```python
import requests
!pip install beautifulsoup4
from bs4 import BeautifulSoup
soup=BeautifulSoup(requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text,'html.parser')
#print(soup.prettify())
```

<h2 style='color:#3186cc'><span style='color:red'>2.</span> Extract table from the HTML page</h2>

```python
my_table=soup.find_all('table',class_='wikitable sortable')
#my_table
```

<h2 style='color:#3186cc'><span style='color:red'>3. </span>Extract data from table and place it in the DataFrame.</h2>

```python
A=[]
B=[]
C=[]
for row in my_table[0].find_all('tr'):
    if len(row.find_all('td'))==3:
        A.append(row.find_all('td')[0].find(text=True))
        B.append(row.find_all('td')[1].find(text=True))
        C.append(row.find_all('td')[2].find(text=True))
df=pd.DataFrame(A,columns=['Postal Code'])
df['Borough']=B
df['Neighborhood']=C
df
```

<h2 style='color:#3186cc'><span style='color:red'>4. </span>Clean up the data and make it look better</h2>

```python
for i in range(180):
    for v in range(3):
        df.values[i,v]=df.values[i,v].split('\n')[0]
df
```

<h2 style='color:#3186cc'><span style='color:red'>5. </span>Replace 'Not assigned' with np.nan to delete the rows if borough is not assigned </h2>

```python
df.replace('Not assigned',np.nan,inplace=True)
df.dropna(subset=['Borough'],axis=0,inplace=True)
df
```

<h2 style='color:#3186cc'><span style='color:red'>6. </span>Check columns if NaN exists</h2>

```python
for col in df.isnull().columns:
    print(col)
    print(df.isnull()[col].value_counts())
    print('')
```

<h2 style='color:#3186cc'><span style='color:red'>7. </span>Reset index and check dataframe shape</h2>

```python
df=df.reset_index(drop=True)
df
```

```python
print(df.shape)
```

<h1 style='color:red;font-size:300%;text-align:center',id='C2'>Part Two</h1>

<h2 style='color:#3186cc'><span style='color:red'>8. </span>Load Geospatial file named df1 and merge df and df1 through .join method</h2>

```python
df1=pd.read_csv('https://cocl.us/Geospatial_data')
df2=df.join(df1.set_index('Postal Code'),on='Postal Code')
df2
```

<h1 style='color:red;font-size:300%;text-align:center';id='C3'>Part Three</h1>

<h2 style='color:#3186cc'><span style='color:red'>9. </span>Explore and cluster the neighborhoods in North York</h2>

<h2 style='color:#3186cc;font-size:20px'> 9.1 Filter data with North York</h2>

```python
df3=df2[df2.Borough=='North York'].reset_index(drop=True)
df3
```

<h2 style='color:#3186cc;font-size:20px'> 9.2 Produce the map of North York</h2>

```python
!conda install -c conda-forge folium=0.5.0 --yes
import folium
latitude=df3.values[0,3]
longitude=df3.values[0,4]
map_northyork=folium.Map([latitude,longitude],zoom_start=11)
for lat,lng,label in zip(df3.Latitude,df3.Longitude,df3.Neighborhood):
    folium.CircleMarker([lat,lng],radius=5,color='blue',popup=label,fill=True,fill_color='blue',fill_opacity=0.5,parse_html=False).add_to(map_northyork)
map_northyork
```

<h2 style='color:#3186cc;font-size:20px'> 9.3 Cluster data with KMeans</h2>

```python
from sklearn.cluster import KMeans
k=3
clustering=pd.concat([df3[['Latitude','Longitude']],pd.get_dummies(df3.Neighborhood,prefix='',prefix_sep='')],axis=1)
KM=KMeans(n_clusters=k).fit(clustering.drop(clustering.columns[[0,1]],axis=1))
clustering.insert(0,'Labels',KM.labels_)
print(clustering.shape)
clustering.head()
```

<h2 style='color:#3186cc;font-size:20px'> 9.4 Visualize the clustering data</h2>

```python
import matplotlib.cm as cm,matplotlib.colors as colors
map_clustering=folium.Map([latitude,longitude],zoom_start=11)
x=np.arange(k)
y=[i+x+(i*x)**2 for i in range(k)]
colors_array=cm.rainbow(np.linspace(0,1,len(y)))
rainbow=[colors.rgb2hex(i) for i in colors_array]
for lat,lng,label in zip(clustering.Latitude,clustering.Longitude,clustering.Labels):
    labe=folium.Popup('Cluster '+str(label),parse_html=True)
    folium.CircleMarker([lat,lng],radius=5,color=rainbow[label-1],popup=labe,fill=True,fill_color=rainbow[label-1],fill_opacity=0.5).add_to(map_clustering)
map_clustering        
```
